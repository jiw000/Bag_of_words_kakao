```python
import nltk
from konlpy.tag import Okt
from tqdm import tqdm
okt = Okt()
```

```python
pos_map = {}
for i in tqdm(data["text"].index):
    text = data.loc[i, "text"]
    poss = okt.pos(text)
    pos_map[i] = poss

pos_map
```

```python
aavn = {}

for i in tqdm(data['text'].index):
    text = pos_map[i]

    aavn[i] = []
    for word, tag in text:
        if tag in ['Noun','Adjective','Verb','Adverb']:
            aavn[i].append(word)

aavn
```

```python
# 단어 중복제거

unique_words = []

for words in tqdm(aavn.values()):
    unique_words.extend(words)
    unique_words = list(set(unique_words))
```

```python
len(unique_words)
```

```python
len(aavn)
```

```python
count = pd.DataFrame(columns = unique_words)
count
```

```python
for i in tqdm(data['text'].index):

    vc = pd.Series(aavn[i]).value_counts()

    count.loc[i, vc.index] = vc.values 


count
```

```python
count = count.fillna(0)
```

```python
count['subject'] = data['subject']
count
```

```python
unique_key = count["subject"].unique()

section_map = {}
section_unmap = {}
for i, subject in enumerate(unique_key):
    section_map[subject] = i
    section_unmap[i] = subject

section_map
```

```python
count["subject"] = count["subject"].replace(section_map)
count
```

### Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# 변수 설정
feature = list(count.columns.difference(['subject']))
x = count[feature]
y = count['subject']

# 학습데이터와 평가데이터의 비율을 8:2 로 분할
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.2, random_state = 42)
print(train_x.shape, test_x.shape, train_y.shape, test_y.shape) 
```

```python
clf = RandomForestClassifier(n_estimators=300,random_state=0)
clf.fit(train_x,train_y)

predict = clf.predict(test_x)
print(accuracy_score(test_y,predict))
```
