```python
# 모든 파일 불러들여서 리스트에 넣기 

DataFrame = []
for json in json_file_list:
    with open(f'/content/drive/MyDrive/Adam_데분동아리/Bag_of_word_pro_01/TL_01. KAKAO(1)/{json}') as f:
        long_description = f.read()
        long_description = ast.literal_eval(long_description) # str을 dict로 변환
        DataFrame.append(long_description)
DataFrame
```

```python
subject = []
text =[]

for c in range(0,2375):
    end = DataFrame[c]['info'][0]['annotations']['lines'][-1]['id'] # 각 주제 대화마다 대화한 사람의 수가 다르기 때문에 총 사람수를 따로 구함 
    for i in range(0,end -1):
        subject.append(DataFrame[c]['info'][0]['annotations']['subject'])
        text.append(DataFrame[c]['info'][0]['annotations']['lines'][i]['norm_text'])
```

```python
result=pd.DataFrame({"subject":subject,
                    "text":text
                    })
```
